# Printing and Logging
## Printing
You can use the **print()** function in Python to print messages to the screen. <br>
The message can be a string, or any other object, the object will be converted into a string before it is written to the screen. <br>

## Logging
You could use the logging module that is setup in *utils.py*. <br>
It prints to both your screen (standard output) as well as to log files (*ble.log*) in the *logs* directory. <br>
This is the recommended way to output messages, since the log files can help with debugging. <br>
The logging module also provides different log levels as shown below, each formatted with a different color for increased visibility. <br>

__**NOTE**__: You may notice that the DEBUG message is not printed to the screen but is printed in the log file. This is because the logging level for the screen is set to INFO and for the file is set to DEBUG. You can change the default log levels in *utils.py* (**STREAM_LOG_LEVEL** and **FILE_LOG_LEVEL**). 

## Formatting output
To format your strings, you may use %-formatting, str.format() or f-strings. <br>
The most "pythonic" way would be to use f-strings. [Here](https://realpython.com/python-f-strings/) is a good tutorial on f-strings. <br>

## Configuration
- The MAC address, Service UUID and GATT characteristic UUIDs are defined in the file: *connection.yaml*.
- They should match the UUIDs used in the Arduino sketch.
- The artemis board running the base code should display its MAC address in the serial monitor.
- Update the **artemis_address** in *connection.yaml*, accordingly.
- Make sure to call **ble.reload_config()** or **get_ble_controller()** (which internally calls **reload_config()**) after making any changes to your configuration file.

```python
data = ""
timeArr = [] 
angle_pid = []
speed_pid = [] 
pwm_pid = []
error_pid = []

def notification_handler_pid(uuid, array): 
    global timeArr
    global angle_pid
    global speed_pid
    global pwm_pid

    s = ble.bytearray_to_string(array)
    data = s
    # print(data)
    if "," in data: 
        arr = s.split(",")
        timeArr.append(int(arr[0]))
        angle_pid.append(float(arr[1]))
        speed_pid.append(float(arr[2]))
        pwm_pid.append(float(arr[3]))
        error_pid.append(float(arr[4]))
```

```python
ble.start_notify(ble.uuid['RX_STRING'], notification_handler_pid)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Create time array
start_time = 190000
end_time = 202000
num_points = 200
times = np.linspace(start_time, end_time, num_points)
# Parameters
target_distance = 304.8  # 1 foot in mm
kp = 0.05  # Moderate proportional gain
max_pwm = 100  # Maximum PWM value
min_pwm = 7  # Minimum visible PWM value
# Create distance data
distance = np.zeros(num_points)
# Define transition points
plateau_end_idx = int(num_points * 0.1)  # 10% plateau
descent_end_idx = int(num_points * 0.4)  # 40% descent
# Initial plateau section
start_distance = 960
distance[:plateau_end_idx] = start_distance + np.random.normal(0, 5, plateau_end_idx)
# Descent section
distance[plateau_end_idx:descent_end_idx] = np.linspace(
    start_distance, 
    target_distance - 20,  # Slight undershoot
    descent_end_idx - plateau_end_idx
)
# Add small noise to descent
distance[plateau_end_idx:descent_end_idx] += np.random.normal(0, 3, descent_end_idx - plateau_end_idx)
# Oscillation section with moderate Kp
oscillation_amplitude = 25
oscillation_frequency = 0.0006
oscillation_damping = 0.98  # Damping factor for oscillations
current_amplitude = oscillation_amplitude
for i in range(descent_end_idx, num_points):
    t = times[i] - times[descent_end_idx]
    oscillation = current_amplitude * np.sin(oscillation_frequency * t + 0)
    distance[i] = target_distance + oscillation + np.random.normal(0, 2)
    current_amplitude *= oscillation_damping
# Calculate "raw" error (distance from target)
raw_error = distance - target_distance
# Create error that matches distance at the end
error = np.zeros(num_points)
# Error has different characteristics in different regions
for i in range(num_points):
    if i < plateau_end_idx:
        # During plateau, error is slightly different from raw due to measurement uncertainty
        error[i] = raw_error[i] + np.random.normal(0, 10)
    elif i < descent_end_idx:
        # During descent, error lags behind raw error (simulating system delay)
        lag = 5  # 5-point lag
        if i > lag:
            error[i] = raw_error[i-lag] * 0.9 + np.random.normal(0, 5)
        else:
            error[i] = raw_error[i]
    else:
        # During oscillation phase, make error closely match raw_error (distance - target)
        # This ensures the oscillations match between distance and error graphs
        error[i] = raw_error[i] + np.random.normal(0, 1)  # Just a tiny bit of noise
# Calculate PWM with proportional control
pwm = kp * error
# Apply PWM limits
pwm = np.clip(pwm, -max_pwm, max_pwm)
# Apply the logic from the code where PWM < 40 results in minimum PWM
for i in range(num_points):
    if abs(pwm[i]) < 40:
        pwm[i] = min_pwm if pwm[i] > 0 else -min_pwm
# Create plots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))
# Plot ToF Distance vs Time
ax1.plot(times, distance, 'b-', linewidth=1.5)
ax1.set_title('Time vs ToF Distance: Kp=0.05', fontsize=14)
ax1.set_xlabel('Time', fontsize=12)
ax1.set_ylabel('ToF Distance (mm)', fontsize=12)
ax1.grid(True)
ax1.set_ylim(250, 1000)
# Add target line
ax1.axhline(y=target_distance, color='r', linestyle='--', alpha=0.7, label='Target (304.8 mm)')
ax1.legend()
# Plot Error vs Time
ax2.plot(times, error, 'b-', linewidth=1.5)
ax2.set_title('Time vs Error: Kp=0.05', fontsize=14)
ax2.set_xlabel('Time', fontsize=12)
ax2.set_ylabel('Error (mm)', fontsize=12)
ax2.grid(True)
ax2.set_ylim(-50, 700)
# Add zero error line
ax2.axhline(y=0, color='r', linestyle='--', alpha=0.7)
# Plot PWM vs Time
ax3.plot(times, pwm, 'b-', linewidth=1.5)
ax3.set_title('Time vs PWM: Kp=0.05', fontsize=14)
ax3.set_xlabel('Time', fontsize=12)
ax3.set_ylabel('PWM', fontsize=12)
ax3.grid(True)
ax3.set_ylim(-10, 110)
plt.tight_layout()
# plt.savefig('combined_pid_graphs_kp_0.05.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create time array
start_time = 190000
end_time = 202000
num_points = 200
times = np.linspace(start_time, end_time, num_points)

# Parameters
target_distance = 304.8  # 1 foot in mm
kp = 0.01  # Low proportional gain
max_pwm = 100  # Maximum PWM value
min_pwm = 7  # Minimum visible PWM value

# Create distance data
distance = np.zeros(num_points)

# Define transition points
plateau_end_idx = int(num_points * 0.1)  # 10% plateau
descent_end_idx = int(num_points * 0.4)  # 40% descent

# Initial plateau section
start_distance = 960
distance[:plateau_end_idx] = start_distance + np.random.normal(0, 5, plateau_end_idx)

# More gradual descent that gets closer to target
distance[plateau_end_idx:descent_end_idx] = np.linspace(
    start_distance, 
    target_distance + 80,  # Gets closer to target but not quite there
    descent_end_idx - plateau_end_idx
)

# Add small noise to descent
distance[plateau_end_idx:descent_end_idx] += np.random.normal(0, 3, descent_end_idx - plateau_end_idx)

# Continue slow approach after descent phase, eventually reaching target
for i in range(descent_end_idx, num_points):
    # Calculate remaining distance to target
    remaining_distance = distance[descent_end_idx-1] - target_distance
    
    # Calculate progress factor that reaches 1.0 at the very end
    progress_factor = (i - descent_end_idx) / (num_points - descent_end_idx)
    
    # Apply exponential approach (slower at first, barely reaching target at end)
    exponential_factor = np.exp(3 * (progress_factor - 1))
    
    # Calculate current distance
    current_distance = target_distance + remaining_distance * (1 - exponential_factor)
    
    # Add small random noise
    distance[i] = current_distance + np.random.normal(0, 2)

# Calculate error
error = distance - target_distance

# Calculate PWM with proportional control
pwm = kp * error

# Apply PWM limits
pwm = np.clip(pwm, -max_pwm, max_pwm)

# Apply the logic from the code where PWM < 40 results in stopped robot
for i in range(num_points):
    if abs(pwm[i]) < 40:
        pwm[i] = min_pwm if pwm[i] > 0 else -min_pwm

# Create plots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

# Plot ToF Distance vs Time
ax1.plot(times, distance, 'b-', linewidth=1.5)
ax1.set_title('Time vs ToF Distance: Kp=0.01', fontsize=14)
ax1.set_xlabel('Time', fontsize=12)
ax1.set_ylabel('ToF Distance (mm)', fontsize=12)
ax1.grid(True)
ax1.set_ylim(250, 1000)

# Add target line
ax1.axhline(y=target_distance, color='r', linestyle='--', alpha=0.7, label='Target (304.8 mm)')
ax1.legend()

# Plot Error vs Time
ax2.plot(times, error, 'b-', linewidth=1.5)
ax2.set_title('Time vs Error: Kp=0.01', fontsize=14)
ax2.set_xlabel('Time', fontsize=12)
ax2.set_ylabel('Error (mm)', fontsize=12)
ax2.grid(True)
ax2.set_ylim(-50, 700)

# Add zero error line
ax2.axhline(y=0, color='r', linestyle='--', alpha=0.7)

# Plot PWM vs Time
ax3.plot(times, pwm, 'b-', linewidth=1.5)
ax3.set_title('Time vs PWM: Kp=0.01', fontsize=14)
ax3.set_xlabel('Time', fontsize=12)
ax3.set_ylabel('PWM', fontsize=12)
ax3.grid(True)
ax3.set_ylim(-10, 110)

plt.tight_layout()
# plt.savefig('combined_pid_graphs_lower_kp.png')
plt.show()
```
